````python
# Here we import all packages we need.

# This is a package for handling and manipulating data.
import pandas as pd

# This is the standard package for plotting.
from matplotlib import pyplot as plt 

# Seaborn is build on matplotlib.
# It is more confortable to use than pure matplotlib.
import seaborn as sns
````

````python
#Loading the dataset
penguins = sns.load_dataset("penguins")
````

### Grouping  
Grouping allows us to split the data according a chosen criteria. We can then calculate the average for each group.  
We can use grouping to find out the average measurements of each penguins species.

````python
#This produces a grouping, based on different kinds of data 
groups = penguins.groupby('species')

#We can get access to each data frame like this:
groups.get_group('Adelie')
````

````python
groups.get_group('Chinstrap')
````

````python
groups.get_group('Gentoo')
````

````python
# Here, we calculate the mean for each group.
# This again produces a single data frame.
species_avg = groups.mean()
species_avg
````

### Plotting data with Seaborn and Matplotlib  

 #### Scatter Plot 
  A scatter plot (aka scatter chart, scatter graph) uses dots to represent values for two different numeric variables.  
The position of each dot on the horizontal and vertical axis indicates values for an individual data point.  
Scatter plots are used to observe relationships between variables.

We will use scatterplot to map 'bill length (mm)' to the x-location, 'bill depth (mm)' to the y-location, and 'species' to the colour hue.

````python
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue = 'species')
````

Adding query to sort only 'Biscone' island penguins and then plot them

````python
# Here we do the same plot, but consider only penguins from Biscoe island:
biscoe=penguins.query('island == "Biscoe"')
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', hue = 'species')
````

#### Controlling the plot appearance  
We can make the plots more applealing to us (humans) by changing color or appearance to help direct our focus onto a specific data.

*   We will now control the colormap in scatterplot function via palette parameter.
*  ```hls``` color space, which is a simple transformation of RGB values
*   Now we will move legend outside the Box to make it more readable.





````python
# Here we additionally map 'body mass (g)' to the size and 'sex' to the style of the markers.
sns.scatterplot(data=penguins, x='bill_length_mm', y='bill_depth_mm', 
                hue = 'species', size = 'body_mass_g', style='island',
# Here we set the colormap to 'hls'
                palette="hls")

# We use this command to move the legend to outside of the plot.
plt.legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.0, frameon = False)
````

#### Bar plots  
It shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar.  
The size of the bar represents its numeric value

````python
# Making a standard vertical barplot
species_avg = species_avg.reset_index()
sns.barplot(data=species_avg, x='species',y='bill_length_mm')
sns.despine()
````

````python
# Making a horizontal barplot
sns.barplot(y='species',x='bill_length_mm', data=species_avg, orient='h')
sns.despine()
````

#### Hist Plot  
  
A histogram is a classic visualization tool that represents the distribution of one or more variables by counting the number of observations that fall within disrete bins


````python
# Directly plot the data for all parameters
sns.histplot(data=penguins)
````

````python
# Set x axis value and set bin number
sns.histplot(data=penguins, x="flipper_length_mm", bins=30)
````

````python
# Plot the KDE on Histogram
# KDE represents the data using a continuous probability density curve in one or more dimensions
sns.histplot(data=penguins, x="flipper_length_mm", kde=True)
````

````python
#We can draw multiple histograms from a long-form dataset with hue mapping
sns.histplot(data=penguins, x="flipper_length_mm", hue="species")
````

````python
#We can also use 'poly' by drawing a polygon with vertices in the center of each bin. This helps get a good visualisation of distribution of data
sns.histplot(penguins, x="flipper_length_mm", hue="species", element="poly")
````

#### Pair Plot  
A pair plot is a matrix of scatterplots that lets us understand the pairwise relationship between different variables in a dataset.


````python
sns.pairplot(penguins)
````

````python
# Add hue parameter
sns.pairplot(penguins, hue="species") 
````

````python
# Set diagonal plot to be overlapping histograms
sns.pairplot(penguins, hue="species", diag_kind="hist")
````

#### Example 1  
Calculate the ratio between `bill length (mm)` and `flipper length (mm)` and add the result as a new column named `bf ratio` to the penguins data frame.

Make a scatterplot mapping `bf ratio` to x-location and `bill ratio` to y-location.

#### Example 2  
Divide the penguins into sub-populations such that each sub-population consists of the penguins that belong to the same species, have the same 'sex', and live on the same 'island'.

For example: All male Adelie penguins on Torgersen would be one sub-population. All female Adelie penguins on Torgersen island would be another sub-population. All male Adelie penguins on Dream island would be yet another sub-population, and so on.

Make a scatterplot in which each data point (marker) corresponds to one such sub-population. Map the average bill length (mm) in the sub-population to the x-location, the average bill depth (mm) to the y-location, the sex of the sub-population to the marker style, and the species to the hue.

Can you additionally map the size (number of penguins) of the sub-population to the marker size? 


````python
island_sex_species = penguins.groupby(['island','sex','species']).mean().reset_index()
island_sex_species
````

````python
sns.scatterplot(data= island_sex_species, x='bill_length_mm', y='bill_depth_mm',
               hue= 'species', style = 'sex'  )
````

````python
island_sex_species = penguins.groupby(['island','sex','species']).mean()
island_sex_species_count = penguins.groupby(['island','sex','species']).count()


island_sex_species['count'] = island_sex_species_count['bill ratio']
island_sex_species= island_sex_species.reset_index()
sns.scatterplot(data= island_sex_species, x='bill_length_mm', y='bill_depth_mm',
               hue= 'species', style = 'sex', size = 'count'  )

plt.legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.0)
````

#### Example 3
Make a single horizontal barplot, showing how each penguin species is distributed across the three islands.
Map `species` to the y-location, the `island` to hue and the penguin `count` to x-location.

````python
island_count=penguins.groupby(['island','species']).count().reset_index()[['island','species', 'bill_length_mm']]
island_count = island_count.rename(columns={'bill_length_mm':'count'})
island_count = island_count.sort_values('count', ascending= True)
sns.barplot(y='species',x='count',hue= 'island', data=island_count,orient='h')
sns.despine()
````
